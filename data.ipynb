```python
import yfinance as yf
import pandas as pd

# Set the start and end date
start_date = '1990-01-01'
end_date = '2023-11-27'

# Set the ticker
ticker = 'EURUSD=X'

# Set the interval to '1m' for 1-minute data
interval = '1m'

# Initialize an empty DataFrame to store the results
all_data = pd.DataFrame()

# Define a function to download data for a specific date range
def download_data_chunk(start, end):
    try:
        return yf.download(ticker, start, end, interval=interval)
    except Exception as e:
        print(f"Failed to download data for {start} - {end}: {e}")
        return pd.DataFrame()

# Loop through the date range in chunks of 7 days
date_chunks = pd.date_range(start=start_date, end=end_date, freq='7D')
for i in range(len(date_chunks) - 1):
    chunk_start = date_chunks[i]
    chunk_end = date_chunks[i + 1]
    
    # Download data for the current chunk
    chunk_data = download_data_chunk(chunk_start, chunk_end)
    
    # Concatenate the data to the main DataFrame
    all_data = pd.concat([all_data, chunk_data])

# Print the result
print(all_data)
```

```python
from forex_python.converter import get_rate
from datetime import datetime, timedelta

# Define the currency pair and time range
base_currency = 'EUR'
target_currency = 'USD'
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 1, 2)

# Set the time interval (1 second in this case)
interval = timedelta(seconds=1)

# Initialize a list to store historical data
historical_data = []

# Fetch historical data at the specified interval
current_date = start_date
while current_date <= end_date:
    try:
        # Get exchange rate at the current date and time
        rate = get_rate(base_currency, target_currency, date_obj=current_date)
        
        # Append data to the list
        historical_data.append({'datetime': current_date, 'rate': rate})
        print(f"{current_date}: {rate}")
    except:
        pass  # Handle exceptions, e.g., if data is not available for a specific timestamp
    
    # Increment the current date by the specified interval
    current_date += interval

# Convert the list to a DataFrame
historical_df = pd.DataFrame(historical_data)

# Print the result
print(historical_df)
```
